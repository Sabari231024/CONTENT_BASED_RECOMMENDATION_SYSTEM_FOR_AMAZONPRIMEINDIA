```python
#content based filtereing
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import NearestNeighbors



df1['episode_count'].fillna(0, inplace=True)
features = ['genre', 'moods', 'content_advisory', 'audio_languages', 'subtitles', 'directors', 'producers', 'starring', 'studio', 'runtime']
X = df1[features]
y = df1['imdb_rating']
numeric_features = ['runtime']
categorical_features = ['genre', 'moods', 'content_advisory', 'audio_languages', 'subtitles', 'directors', 'producers', 'starring', 'studio']
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
X_processed = preprocessor.fit_transform(X)
k = 8000  
knn = NearestNeighbors(n_neighbors=k)
knn.fit(X_processed)


movie_index = 0  #this depends on latest movie watched by user
query_movie = X_processed[movie_index].reshape(1, -1)
distances, indices = knn.kneighbors(query_movie)
similar_movies_indices = indices[0][1:] 
similar_movies_distances = distances[0][1:] 
similar_movies = df1.iloc[similar_movies_indices]
similar_movies['distance'] = similar_movies_distances
print(similar_movies[['name', 'distance']])
```
